# Question 1:
Please write a program using generator to print the numbers which can be divisible by 5 and 7 between 0 and n in comma separated form while n is input by console.
Example:

```python
import logging as lg

# input variable
n = int(input('enter any number : '))

# iter object
number = iter(range(n))
dummy = 0

# set up logging
lg.basicConfig(format='%(asctime)s - %(levelname)s - %(message)s - %(name)s',
                level=lg.DEBUG,
                handlers=[lg.FileHandler('variable_1.log'), lg.StreamHandler()])

# create logger object
logger = lg.getLogger('test')

# loop
while dummy < n:
    dummy += 1
    a = next(number)  # generator
    if a % 5 == 0 and a % 7 == 0:  # condition check
        logger.info(a)

    if a >= dummy:  # break the loop
        break
```
